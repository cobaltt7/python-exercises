# Shapiro-Wilk test

- Tests whether a sample comes from a normal distrubtion
- H0 - Population is distributed normally
- H1 - Population has an abnormal distribution


```python
from numpy.random import randn

# Create data
gfg_data = randn(500)
gfg_data
```

```python
from scipy.stats import shapiro

_, pvalue = shapiro(gfg_data)
pvalue
```

```python
"Non-normal" if pvalue > 0.05 else "Normal"
```

```python
import matplotlib.pyplot as plt

plt.hist(gfg_data)
```

```python
from numpy.random import poisson

# Create data
gfg_data = poisson(5, 200)
gfg_data
```

```python
_, pvalue = shapiro(gfg_data)
pvalue
```

```python
"Non-normal" if pvalue > 0.05 else "Normal"
```

```python
plt.hist(gfg_data)
```
