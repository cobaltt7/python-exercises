# Strip plot

```python
import seaborn

tips = seaborn.load_dataset("tips")
tips
```

```python
import matplotlib.pyplot as plt

seaborn.set(style="whitegrid")
seaborn.stripplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
seaborn.stripplot(x=tips["total_bill"])
```

```python
seaborn.stripplot(x="day", y="total_bill", data=tips, jitter=0.1)
```

```python
seaborn.stripplot(y="total_bill", x="day", data=tips, linewidth=3)
```

```python
seaborn.stripplot(y="total_bill", x="day", data=tips, linewidth=2, edgecolor="green")
```

```python
seaborn.stripplot(x="sex", y="total_bill", hue="day", data=tips)
```

```python
seaborn.stripplot(
    x="day", y="total_bill", hue="smoker", data=tips, palette="Set2", dodge=True
)
```

```python
seaborn.stripplot(
    x="day",
    y="total_bill",
    hue="smoker",
    data=tips,
    palette="Set1",
    size=20,
    marker="s",
    alpha=0.2,
)

plt.show()
```
