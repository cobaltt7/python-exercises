# Pair Plot
- Matrix of scatter plots comparing every column with every column
- (Instead of comparing columns with themselves, it shows another graph; histogram by default)

```python
import seaborn

df = seaborn.load_dataset("tips")
df
```

```python
import matplotlib.pyplot as plt
seaborn.pairplot(df)
plt.show()
```

```python
selected_vars = ["total_bill", "tip"]
seaborn.pairplot(df, vars=selected_vars)
plt.show()
```

```python
seaborn.pairplot(df, hue="size")
plt.show()
```

```python
seaborn.pairplot(df, hue="size", palette="husl")
plt.show()
```

```python
seaborn.pairplot(df, diag_kind="kde")
plt.show()
```

```python
seaborn.pairplot(df, kind="reg")
plt.show()
```

```python
seaborn.pairplot(df, hue="sex", markers=["o", "s"])
plt.show()
```

```python
seaborn.pairplot(df, hue="sex", vars=["total_bill", "tip", "size"])
plt.show()
```

```python
g = seaborn.pairplot(df, hue="day")
g.fig.suptitle("Pairplot of Tips Dataset", y=1.02)  # Add a title
g.set(xticks=[], yticks=[])  # Remove tick labels
plt.show()
```
