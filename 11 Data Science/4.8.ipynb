# One-Way ANOVA
- Analyis of variance
- Compares the means of 3+ groups
- H0 (null hypothesis) All means are equal
- H1: At least one mean is different

```python
performance1 = [89, 89, 88, 78, 79]
performance2 = [93, 92, 94, 89, 88]
performance3 = [89, 88, 89, 93, 90]
performance4 = [81, 78, 81, 92, 82]
```

```python
%conda install scipy
```

```python
from scipy.stats import f_oneway

f_statistic, p_value = f_oneway(performance1, performance2, performance3, performance4)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

- F-stat: variance between groups vs variance within groups
- P-val: To determine if the results are statistically significant

`p-val < 0.05` -> reject H0 -> there is a difference
