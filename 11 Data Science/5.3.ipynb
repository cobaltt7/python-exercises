```python
import numpy as np

mean, stdDev = 121, 21
x = np.random.normal(mean, stdDev, 1000)
x
```

```python
n, bins, _ = plt.hist(x, bins=100, density=True, color="green", alpha=0.7)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("matplotlib.pyplot.hist() Example", fontweight="bold")

plt.show()
```

```python
x = np.random.randn(10000, 3)
colors = ["green", "blue", "lime"]

plt.hist(x, bins=20, density=True, histtype="bar", color=colors, label=colors)

plt.legend(fontsize=10)
plt.title("matplotlib.pyplot.hist() Example", fontweight="bold")

plt.show()
```
