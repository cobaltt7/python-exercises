## Univariate Analysis
One variable

```python
quality_counts = df["quality"].value_counts()
quality_counts
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color="deeppink")
plt.title("Count Plot of Quality")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

### Kernel density plot

```python
import seaborn as sns
sns.set_style("darkgrid")
```

```python
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns
numerical_columns
```

```python
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

- Skewness of 0 = perfectly symmetrical
- \>0 is right-skewed
- <0 is left-skewed

### Swarm Plot

```python
plt.figure(figsize=(15, 12))

sns.swarmplot(x="quality", y="alcohol", data=df, palette="viridis", hue="quality")

plt.title("Swarm Plot for Quality and Alcohol")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.show()
```

## Bivariate Analysis
Two variables

### Pair Plot
Lattice of all possible scatter plots, searching for correlation

```python
sns.set_palette("Pastel1")

plt.figure(figsize=(10, 6))

sns.pairplot(df)

plt.suptitle("Pair Plot for DataFrame")
plt.show()
```

### Violin plot
(Alcohol and Qualiy here)
- Wider width = higher density
- Symetrical plots have a balanced distribution
- Peaks or buldges show common values
- Longer tails show greater variability
- Median is the middle line

```python
df["quality"] = df["quality"].astype(str)

plt.figure(figsize=(10, 8))

sns.violinplot(
    x="quality",
    y="alcohol",
    data=df,
    palette={
        "3": "lightcoral",
        "4": "lightblue",
        "5": "lightgreen",
        "6": "gold",
        "7": "lightskyblue",
        "8": "lightpink",
    },
    alpha=0.7,
    hue="quality",
)

plt.title("Violin Plot for Quality and Alcohol")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.show()
```

### Box plot
- Median line inside the box
- Whiskers from the box to the end of the IQRs
- Points for outliers
- Compact boxes show less variability

```python
sns.boxplot(x="quality", y="alcohol", data=df)
plt.show()
```

## Multivariate Analysis

## Correlation matrix plot
- 1 = Strong positive correlation (increase together)
- 0 = No correlation
- -1 = Strong negative correlation (one increases, other decreases)
- Darker colors represent stronger correlation

```python
plt.figure(figsize=(15, 10))

sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="Pastel2", linewidths=2)

plt.title("Correlation Heatmap")
plt.show()
```
