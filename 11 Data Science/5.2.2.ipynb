# Count Plots
Aka Multiple bar plots

```python
import seaborn as sns

df = sns.load_dataset("tips")
df
```

```python
import matplotlib.pyplot as plt

sns.countplot(x="sex", data=df)
plt.show()
```

```python
sns.countplot(x="sex", hue="smoker", data=df)
plt.show()
```

```python
sns.countplot(y="sex", hue="smoker", data=df)
plt.show()
```

```python
sns.countplot(x="sex", data=df, palette="Set2", hue="sex")
plt.show()
```

```python
df = sns.load_dataset("titanic")
df
```

```python
sns.countplot(x="class", hue="sex", data=df, color="salmon")
plt.show()
```

```python
sns.countplot(x="sex", data=df, color="salmon", saturation=0.1)
plt.show()
```

```python
sns.countplot(
    x="sex",
    data=df,
    color="salmon",
    facecolor=(0, 0, 0, 0),
    linewidth=5,
    edgecolor=sns.color_palette("BrBG", 2),
)

plt.show()
```
