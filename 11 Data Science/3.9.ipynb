### Z-Score
- Helps detect outliers
- Counts the number of standard deviations an item is from the mean
- `Z = (X - M) / D` where `Z` is the Z-score, `X` is the item, `M` is the mean, and `D` is the standard deviation
- A Z-score `> 3` or `< -3` is generally considered an outlier

```python
import pandas as pd
df = pd.DataFrame([1, 2, 2, 2, 3, 1, 1, 15, 2, 2, 2, 3, 1, 1, 2], columns=["Value"])
df
```

```python
from scipy.stats import zscore
df["Z-score"] = zscore(df["Value"])
df
```

```python
outliers = df[df["Z-score"].abs() > 3]
outliers
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 6))
plt.scatter(df["Value"], np.zeros_like(df["Value"]), color="blue", label="Data points")
plt.scatter(
    outliers["Value"], np.zeros_like(outliers["Value"]), color="red", label="Outliers"
)
plt.title("Outlier Detection using Z-Score")
plt.xlabel("Value")
plt.legend()
plt.show()
```
