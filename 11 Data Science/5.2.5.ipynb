# Kernal Density Estimate (KDE) Plot

```python
%matplotlib inline
```

```python
from sklearn import datasets
iris = datasets.load_iris()
iris
```

```python
import pandas as pd

iris_df = pd.DataFrame(
    iris.data, columns=["Sepal_Length", "Sepal_Width", "Petal_Length", "Petal_Width"]
)
iris_df
```

```python
iris_df["Target"] = iris.target

iris_df["Target"] = (
    iris_df["Target"]
    .replace([0], "Iris_Setosa")
    .replace([1], "Iris_Vercicolor")
    .replace([2], "Iris_Virginica")
)
iris_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(
    iris_df.loc[(iris_df["Target"] == "Iris_Virginica"), "Sepal_Length"],
    color="b",
    fill=True,
    label="Iris_Virginica",
)

plt.xlabel("Sepal Length")
plt.ylabel("Probability Density")
```

```python
sns.kdeplot(
    iris_df.loc[(iris_df["Target"] == "Iris_Setosa"), "Sepal_Length"],
    color="r",
    fill=True,
    label="Iris_Setosa",
)

sns.kdeplot(
    iris_df.loc[(iris_df["Target"] == "Iris_Virginica"), "Sepal_Length"],
    color="b",
    fill=True,
    label="Iris_Virginica",
)

plt.xlabel("Sepal Length")
plt.ylabel("Probability Density")
```

```python
# Setting up the samples
iris_setosa = iris_df.query("Target=='Iris_Setosa'")
iris_virginica = iris_df.query("Target=='Iris_Virginica'")

# Plotting the KDE Plot
sns.kdeplot(
    x=iris_setosa["Sepal_Length"],
    y=iris_setosa["Sepal_Width"],
    color="r",
    fill=True,
    label="Iris_Setosa",
    cmap="Reds",
    thresh=False,
)
```

```python
sns.kdeplot(
    x=iris_setosa["Sepal_Length"],
    y=iris_setosa["Sepal_Width"],
    color="r",
    fill=False,
    label="Iris_Setosa",
    cmap="Reds",
    thresh=False,
)

sns.kdeplot(
    x=iris_virginica["Sepal_Length"],
    y=iris_virginica["Sepal_Width"],
    color="b",
    fill=False,
    label="Iris_Virginica",
    cmap="Blues",
    thresh=False,
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Bivariate Seaborn KDE Plot")
plt.legend()
plt.show()
```
