```python
import seaborn as sns

df = sns.load_dataset("tips")
df
```

```python
import matplotlib.pyplot as plt

heatmap_data = df.pivot_table(index="day", columns="sex", values="tip", aggfunc="mean")
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu")
plt.show()
```

```python
import numpy as np

data = np.random.random((12, 12))
data
```

```python
plt.imshow(data, cmap="autumn")
plt.title("Heatmap with different color")
plt.show()
```

```python
plt.imshow(data, cmap="autumn", interpolation="nearest")
plt.colorbar()
plt.title("Heatmap with color bar")
plt.show()
```

```python
import matplotlib.colors as colors

# Create a custom color map
# with blue and green colors
colors_list = ["#0099ff", "#33cc33"]
cmap = colors.ListedColormap(colors_list)
cmap
```

```python
data = np.random.randint(0, 100, size=(8, 8))
data
```

```python
plt.imshow(data, cmap=cmap, vmin=0, vmax=100, extent=[0, 8, 0, 8])
for i in range(8):
    for j in range(8):
        plt.annotate(
            str(data[i][j]),
            xy=(j + 0.5, i + 0.5),
            ha="center",
            va="center",
            color="white",
        )

# Add colorbar
cbar = plt.colorbar(ticks=[0, 50, 100])
cbar.ax.set_yticklabels(["Low", "Medium", "High"])

# Set plot title and axis labels
plt.title("Customized heatmap with annotations")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
```

```python
hm = sns.heatmap(data=data, annot=True)

# displaying the plotted heatmap
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("./data/gold_price_data.csv")

# Calculate correlation between columns
corr_matrix = df.corr()
corr_matrix
```

```python
# Plot the heatmap with custom colors and annotations
plt.imshow(corr_matrix, cmap=cmap, vmin=0, vmax=1, extent=[0, 5, 0, 5])
for i in range(5):
    for j in range(5):
        plt.annotate(
            str(round(corr_matrix.values[i][j], 2)),
            xy=(j + 0.25, i + 0.7),
            ha="center",
            va="center",
            color="white",
        )

# Add colorbar
cbar = plt.colorbar(ticks=[0, 0.5, 1])
cbar.ax.set_yticklabels(["Low", "Medium", "High"])

# Set plot title and axis labels
plt.title("Correlation Matrix Of The Dataset")
plt.xlabel("Features")
plt.ylabel("Features")

# Set tick labels
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

# Display the plot
plt.show()
```
