# Correlation
- A measure of how closely two variables are related
- Or, how well you can predict one variable knowing the other
- Correlation is generally considered high when `|correlation| >= 0.6`

```python
import pandas as pd

data = pd.DataFrame({"Array_1": [30, 70, 100], "Array_2": [65.1, 49.50, 30.7]})
data
```

```python
data.corr()
```

```python
df = pd.read_csv("./data/nba.csv")
df
```

```python
df[["Number", "Age", "Weight", "Salary"]].corr(method="pearson")
```

```python
df[["Number", "Age", "Weight", "Salary"]].corr(method="kendall")
```
