## Violin Plot

```python
import seaborn

seaborn.set(style="whitegrid")
fmri = seaborn.load_dataset("fmri")
fmri
```

```python
seaborn.violinplot(x="timepoint", y="signal", data=fmri)
```

```python
seaborn.violinplot(x="timepoint", y="signal", hue="region", data=fmri)
```

```python
tip = seaborn.load_dataset("tips")
seaborn.violinplot(x="day", y="tip", data=tip, hue="day")
```

```python
seaborn.violinplot(x=tip["total_bill"])
```

```python
seaborn.violinplot(x="tip", y="day", data=tip)
```

```python
seaborn.violinplot(x="day", y="total_bill", hue="time", data=tip)
```

```python
seaborn.violinplot(x="day", y="tip", data=tip, linewidth=4)
```

```python
seaborn.violinplot(
    x="day", y="total_bill", hue="smoker", data=tip, palette="Set2", dodge=True
)
```

```python
seaborn.violinplot(x="time", y="tip", data=tip, order=["Dinner", "Lunch"])
```

```python
seaborn.violinplot(x="day", y="total_bill", data=tip, hue="time", palette="pastel")
```

```python
seaborn.violinplot(x="day", y="tip", data=tip, saturation=0.03)
```

```python
seaborn.violinplot(x="day", y="tip", data=tip, color="Yellow")
```

```python
seaborn.violinplot(
    x="day",
    y="total_bill",
    hue="sex",
    data=tip,
    palette="Set2",
    split=True,
    density_norm="count",
)
```
