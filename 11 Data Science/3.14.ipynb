```python
import pandas as pd
data = pd.read_csv("./data/employee_data.csv")
data
```

## Univariate Analysis
Analyzing one variable independently
### Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data["age"])
plt.show()
```

### Bar Chart

```python
sns.countplot(data["gender_full"])
plt.show()
```

### Pie Chart

```python
x = data["STATUS_YEAR"].value_counts()
plt.pie(x.values, labels=x.index, autopct="%1.1f%%")
plt.show()
```

## Bivariate Analysis
- Simultaneous analysis of two variables & their relationshp
- Determines if an association exists and its strength
### Categorical vs Numerical
The bar shows the average value and the length of the middle line shows the range

```python
plt.figure(figsize=(40, 20))
sns.barplot(x=data["department_name"], y=data["length_of_service"])
plt.show()
```

### Numerical vs Numerical
Simple scatter plot that clearly shows a relationship (older workers have worked for longer)

```python
sns.scatterplot(x=data["length_of_service"], y=data["age"])
plt.show()
```

### Categorical vs Categorical
Dual bar graph

```python
sns.countplot(data, x="STATUS_YEAR", hue="STATUS")
plt.show()
```

## Multivariate Analysis
Extension of bivariate analysis but with more than 2 variables
### Principle Component Analysis

```python
from sklearn import datasets

iris = datasets.load_iris()
iris
```

```python
from sklearn import decomposition

pca = decomposition.PCA(n_components=2)
pca
```

```python
X = pca.fit_transform(iris.data)
X
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=iris.target)
plt.show()
```

### Heat Map

```python
sns.heatmap(
    data[
        ["EmployeeID", "age", "length_of_service", "store_name", "STATUS_YEAR"]
    ].corr(),
    annot=True,
)
plt.show()
```
