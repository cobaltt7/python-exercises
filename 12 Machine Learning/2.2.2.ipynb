```python
import numpy as np

X = np.array([0.2, 0.4, 0.6])
X
```

```python
y = np.array([0, 1, 1])
y
```

```python
theta = 0.5
z = X * theta
z
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


y_pred = sigmoid(z)
y_pred
```

```python
y_pred
```

```python
def log_loss(y_true, y_pred, eps=1e-15):
    y_pred = np.clip(y_pred, eps, 1 - eps)
    m = y_true.shape[0]
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

log_loss(y, y_pred)
```
